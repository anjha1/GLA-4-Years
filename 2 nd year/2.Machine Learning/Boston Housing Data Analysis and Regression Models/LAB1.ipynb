```python
#@title Outlier management
import numpy as np

# Sample data with outliers
data = {'Age': [25, 30, 120, 40, 50],
        'Salary': [50000, 2000000, 60000, 80000, 70000]}

df = pd.DataFrame(data)

# Defining a function to treat outliers using Z-score
def remove_outliers(df, column):
    mean = df[column].mean()
    std_dev = df[column].std()
    z_score = (df[column] - mean) / std_dev
    return df[np.abs(z_score) < 3]  # Keeping values within 3 standard deviations

# Applying outlier treatment
df = remove_outliers(df, 'Age')
df = remove_outliers(df, 'Salary')

print(df)
```

```python
#@title One hot Encoding
from sklearn.preprocessing import OneHotEncoder

# Sample data with categorical variables
data = {'City': ['New York', 'Paris', 'New York', 'Berlin']}

df = pd.DataFrame(data)

# Using OneHotEncoder
encoder = OneHotEncoder(sparse=False)
one_hot = pd.DataFrame(encoder.fit_transform(df[['City']]),
                       columns=encoder.get_feature_names_out(['City']))

# Joining the one-hot encoded columns back to the original dataframe
df = df.join(one_hot)
print(df)
```

```python
#@title Label Encoding
from sklearn.preprocessing import LabelEncoder

# Using LabelEncoder
label_encoder = LabelEncoder()
df['City_Label'] = label_encoder.fit_transform(df['City'])

print(df)
```

```python
#@title Feature Transformation or Scailing
from sklearn.preprocessing import StandardScaler

# Sample data for scaling
data = {'Age': [25, 30, 35, 40, 50],
        'Salary': [50000, 60000, 70000, 80000, 90000]}

df = pd.DataFrame(data)

# Using StandardScaler to standardize features
scaler = StandardScaler()
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])

print(df)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score

# Initialize individual models
log_clf = LogisticRegression(max_iter=1000)
dt_clf = DecisionTreeClassifier()
knn_clf = KNeighborsClassifier()

# Create a Voting Classifier (hard voting)
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('dt', dt_clf), ('knn', knn_clf)],
    voting='hard'  # Use 'hard' for majority voting
)

# Train the voting classifier
voting_clf.fit(X_train, y_train)

# Predictions
y_pred = voting_clf.predict(X_test)

# Evaluate the accuracy of the voting classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Voting Classifier Accuracy: {accuracy:.4f}")

# Individual classifier performance for comparison
for clf in (log_clf, dt_clf, knn_clf):
    clf.fit(X_train, y_train)
    y_pred_individual = clf.predict(X_test)
    print(f"{clf.__class__.__name__} Accuracy: {accuracy_score(y_test, y_pred_individual):.4f}")
```
